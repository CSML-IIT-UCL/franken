# Minimal Example

Franken models can be easily trained using the autotune CLI tool:

This will create a folder `run_DATE_TIME_...` containing:
* `best_ckpt.pt`  -->  model checkpoint
* `best.json`  -->  train/val/test metrics for the best model
* `config.json`  -->  training configuration
* `franken.log`  -->  stdout from log
* `log.json`  -->  metrics for all tested models (in case of hyperparameter optimization:)

In particular below is an example `best.json` file, which contains info about the **metrics**, **timings**, and **hyperparameters**.

See the `Franken CLI Reference > Autotune` page to view the description of the options and the [dedicated tutorial](./autotune.ipynb).
